```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
import warnings
warnings.filterwarnings('ignore')

%matplotlib inline

# Load cleaned data
train = pd.read_csv('../data/processed/train_cleaned.csv')
test = pd.read_csv('../data/processed/test_cleaned.csv')

print("Train shape:", train.shape)
print("Test shape:", test.shape)
```

```python
# Total square footage of entire house
train['TotalSF'] = train['TotalBsmtSF'] + train['1stFlrSF'] + train['2ndFlrSF']
test['TotalSF'] = test['TotalBsmtSF'] + test['1stFlrSF'] + test['2ndFlrSF']

# Verify
print("TotalSF min:", train['TotalSF'].min())
print("TotalSF max:", train['TotalSF'].max())
print("TotalSF mean:", train['TotalSF'].mean())

# Check correlation with SalePrice
print("\nCorrelation with SalePrice:", train['TotalSF'].corr(train['SalePrice']))
```

```python
# Age of house when sold
train['HouseAge'] = train['YrSold'] - train['YearBuilt']
test['HouseAge'] = test['YrSold'] - test['YearBuilt']

# Years since last remodel
train['RemodAge'] = train['YrSold'] - train['YearRemodAdd']
test['RemodAge'] = test['YrSold'] - test['YearRemodAdd']

# Check correlations
print("HouseAge correlation with SalePrice:", train['HouseAge'].corr(train['SalePrice']))
print("RemodAge correlation with SalePrice:", train['RemodAge'].corr(train['SalePrice']))
```

```python
# Combine all bathrooms into one feature
train['TotalBathrooms'] = (train['FullBath'] + 
                           train['BsmtFullBath'] + 
                           (train['HalfBath'] * 0.5) + 
                           (train['BsmtHalfBath'] * 0.5))

test['TotalBathrooms'] = (test['FullBath'] + 
                          test['BsmtFullBath'] + 
                          (test['HalfBath'] * 0.5) + 
                          (test['BsmtHalfBath'] * 0.5))

print("TotalBathrooms correlation with SalePrice:", train['TotalBathrooms'].corr(train['SalePrice']))
print(train['TotalBathrooms'].value_counts().head())
```

```python
# Combine all porch areas into one feature
train['TotalPorch'] = (train['OpenPorchSF'] + 
                       train['EnclosedPorch'] + 
                       train['3SsnPorch'] + 
                       train['ScreenPorch'] +
                       train['WoodDeckSF'])

test['TotalPorch'] = (test['OpenPorchSF'] + 
                      test['EnclosedPorch'] + 
                      test['3SsnPorch'] + 
                      test['ScreenPorch'] +
                      test['WoodDeckSF'])

print("TotalPorch correlation with SalePrice:", train['TotalPorch'].corr(train['SalePrice']))
```

```python
 # Binary features - 1 if exists, 0 if not
train['HasPool'] = (train['PoolArea'] > 0).astype(int)
train['HasGarage'] = (train['GarageArea'] > 0).astype(int)
train['HasFireplace'] = (train['Fireplaces'] > 0).astype(int)
train['HasBasement'] = (train['TotalBsmtSF'] > 0).astype(int)

test['HasPool'] = (test['PoolArea'] > 0).astype(int)
test['HasGarage'] = (test['GarageArea'] > 0).astype(int)
test['HasFireplace'] = (test['Fireplaces'] > 0).astype(int)
test['HasBasement'] = (test['TotalBsmtSF'] > 0).astype(int)

# Check correlations
binary_features = ['HasPool', 'HasGarage', 'HasFireplace', 'HasBasement']
for col in binary_features:
    print(f"{col} correlation with SalePrice: {train[col].corr(train['SalePrice']):.3f}")
```

```python
# Quality × Condition combined score
train['QualityScore'] = train['OverallQual'] * train['OverallCond']
test['QualityScore'] = test['OverallQual'] * test['OverallCond']

print("QualityScore correlation:", train['QualityScore'].corr(train['SalePrice']))
```

```python
# Living area relative to lot size
train['LotRatio'] = train['GrLivArea'] / train['LotArea']
test['LotRatio'] = test['GrLivArea'] / test['LotArea']

print("LotRatio correlation:", train['LotRatio'].corr(train['SalePrice']))
```

```python
train = train.drop(columns=['LotRatio'])
test = test.drop(columns=['LotRatio'])

print("LotRatio dropped!")
```

```python
# Was the house remodeled after it was built?
train['IsRemodeled'] = (train['YearRemodAdd'] > train['YearBuilt']).astype(int)
test['IsRemodeled'] = (test['YearRemodAdd'] > test['YearBuilt']).astype(int)

print("IsRemodeled correlation:", train['IsRemodeled'].corr(train['SalePrice']))
print(train['IsRemodeled'].value_counts())
```

```python
# Was the house built after 2000?
train['IsNewHouse'] = (train['YearBuilt'] > 2000).astype(int)
test['IsNewHouse'] = (test['YearBuilt'] > 2000).astype(int)

print("IsNewHouse correlation:", train['IsNewHouse'].corr(train['SalePrice']))
print(train['IsNewHouse'].value_counts())
```

```python
# Garage quality combined with size
train['GarageScore'] = train['GarageCars'] * train['GarageArea']
test['GarageScore'] = test['GarageCars'] * test['GarageArea']

print("GarageScore correlation:", train['GarageScore'].corr(train['SalePrice']))
```

```python
# PricePerRoom - living area per room
train['AvgRoomSize'] = train['GrLivArea'] / train['TotRmsAbvGrd']
test['AvgRoomSize'] = test['GrLivArea'] / test['TotRmsAbvGrd']

# Total outdoor area
train['TotalOutdoorSF'] = (train['WoodDeckSF'] + train['OpenPorchSF'] + 
                           train['EnclosedPorch'] + train['ScreenPorch'])
test['TotalOutdoorSF'] = (test['WoodDeckSF'] + test['OpenPorchSF'] + 
                          test['EnclosedPorch'] + test['ScreenPorch'])

# Overall quality squared - amplifies differences between high and low quality
train['QualSquared'] = train['OverallQual'] ** 2
test['QualSquared'] = test['OverallQual'] ** 2

# Total finished basement area
train['TotalFinishedBsmt'] = train['BsmtFinSF1'] + train['BsmtFinSF2']
test['TotalFinishedBsmt'] = test['BsmtFinSF1'] + test['BsmtFinSF2']

# Check all correlations at once
new_features = ['BasementScore', 'AvgRoomSize', 'TotalOutdoorSF', 
                'QualSquared', 'TotalFinishedBsmt']

print("New feature correlations:")
for col in new_features:
    print(f"{col}: {train[col].corr(train['SalePrice']):.3f}")
```

```python
# Verify all new features
all_new_features = ['QualityScore', 'IsRemodeled', 'IsNewHouse', 
                    'GarageScore', 'BasementScore', 'AvgRoomSize',
                    'TotalOutdoorSF', 'QualSquared', 'TotalFinishedBsmt']

print("All new features summary:")
for col in all_new_features:
    print(f"{col}: {train[col].corr(train['SalePrice']):.3f}")

print("\nTrain shape:", train.shape)
print("Test shape:", test.shape)

# Save
train.to_csv('./data/processed/train_featured.csv', index=False)
test.to_csv('./data/processed/test_featured.csv', index=False)

print("\nFeature engineered data saved!")
```

```python
os.makedirs('../data/processed', exist_ok=True)

train.to_csv('./data/processed/train_featured.csv', index=False)
test.to_csv('../data/processed/test_featured.csv', index=False)

print("Feature engineered data saved!")
print("Train shape:", train.shape)
print("Test shape:", test.shape)
```
